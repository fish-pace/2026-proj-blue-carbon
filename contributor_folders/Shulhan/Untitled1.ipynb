```python
import random
from shapely.geometry import Point

def random_points_within(poly, n):
    minx, miny, maxx, maxy = poly.bounds
    points = []

    while len(points) < n:
        x = random.uniform(minx, maxx)
        y = random.uniform(miny, maxy)
        p = Point(x, y)
        if poly.contains(p):
            points.append(p)

    return points
```

```python
points = random_points_within(gdf, 100)
```
